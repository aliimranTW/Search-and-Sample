```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

%matplotlib inline

filename = 'sample.jpg'
image = mpimg.imread(filename)
plt.imshow(image)
plt.show()
```

```python
import numpy as np
print(image.dtype, image.shape, np.max(image), np.min(image))
```

```python
red_channel = np.copy(image)
green_channel = np.copy(image)
blue_channel = np.copy(image)

red_channel[:,:,[1,2]] = 0
green_channel[:,:,[0,2]] = 0
blue_channel[:,:,[0,1]] = 0

fig = plt.figure(figsize = (12,3))
plt.subplot(131)
plt.imshow(red_channel)

plt.subplot(132)
plt.imshow(green_channel)

plt.subplot(133)
plt.imshow(blue_channel)

plt.show()
```

```python
def rgb_thresh(img, thresh = (0,0,0)):
    combined_array = np.zeros_like(img[:,:,0])
    
    for i in range(0,160):
        for j in range(0,320):
            if img[i,j,2] > thresh[2]: #and img[i,j,1] > thresh[1] and img[i,j,2] > thresh[2]:
                combined_array[i,j] = 1
            else:
                combined_array[i,j] = 0
    
    plt.imshow(combined_array)
    plt.show()
    
```

```python
rgb_thresh(image, (160,160,165))
```

```python
def color_thresh(img, rgb_thresh=(0, 0, 0)):
    # Create an array of zeros same xy size as img, but single channel
    color_select = np.zeros_like(img[:,:,0])
    # Require that each pixel be above all three threshold values in RGB
    # above_thresh will now contain a boolean array with "True"
    # where threshold was met
    above_thresh = (img[:,:,0] > rgb_thresh[0]) \
                & (img[:,:,1] > rgb_thresh[1]) \
                & (img[:,:,2] > rgb_thresh[2])
    # Index the array of zeros with the boolean array and set to 1
    color_select[above_thresh] = 1
    
    plt.imshow(color_select)
    plt.show() 
```

```python
color_thresh(image,(160,160,160))
```
